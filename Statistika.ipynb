```python
# Jumlah Kelas

import math 
def jumlahKelas(n):
    kelas = 1 + 3.322 * math.log(n, 10)
    print(kelas)

jumlahKelas(80)
```

```python
# Panjang Interval Kelas

import math
def panjangKelas(maks, mini, kel):
    panjang = (maks - mini) / kel
    print(panjang, '~~~', math.ceil(panjang))

panjangKelas(99, 35, 7)
```

```python
# Frekuensi Kumulatif

def fK(f):
    a = []
    print("fk kurang dari")
    print(0)
    for i in f:
        a.append(i)
        print(sum(a))
    print()
    b = []
    print("fk lebih dari (tulis dari bawah cok)")
    print(0)
    for i in reversed(f):
        b.append(i)
        print(sum(b))
        
# F ITU FREKUENSI TINGGAL GANTI SESUAI SOAL
# INGET BUAT FK LEBIH DARI, TULIS DARI BAWAH

f = [1,2,17,3,10,7]
fK(f)
```

```python
# STANDAR DEVIASI DATA BERKELOMPOK

import math
def standarDev(nilai, frekuensi):
    # menghitung titik tengah
    print("Titik tengah (xi)")
    xi_data = []
    for i in nilai:
        x = (i[0] + i[1]) / 2
        xi_data.append(x)
    print(xi_data)
    print()

    # menghitung f.xi
    print("f.xi")
    fxi_data = []
    j = 0
    while j < len(nilai):
        fxi = xi_data[j] * frekuensi[j]
        j += 1
        fxi_data.append(fxi)
    print(fxi_data)
    print()

    # menghitung rata2
    print("rata-rata (X)")
    rata = sum(fxi_data) / sum(frekuensi)
    print("X =", rata)
    print()

    # menghitung xi - X
    print("xi-X")
    xix_data = []
    for i in xi_data:
        xix = i - rata
        xix_data.append(xix)
    print(xix_data)
    print()

    # menghitung (xi-X)^2
    print("(xi-X)^2")
    xix2data = []
    for i in xix_data:
        xix2 = i ** 2
        xix2data.append(xix2)
    print(xix2data)
    print()

    # menghitung f.(xi-X)^2
    print("fi(xi-X)^2")
    fxix2data = []
    for i in range(len(frekuensi)):
        fxix2 = frekuensi[i] * xix2data[i]
        fxix2data.append(fxix2)
    print(fxix2data)
    print("\n Jumlah f(xi-X)^2 = ", sum(fxix2data))
    print()

    # varian dan standar deviasi
    varian = sum(fxix2data) / sum(frekuensi)
    print("Varian(s^2) =",sum(fxix2data),"/",sum(frekuensi),"=", varian)
    print()
    sb = math.sqrt(varian)
    print("Standar Deviasi (s) = √"+str(varian)+" = "+str(sb))    
        
        
# PANJANG INTERVAL KELAS MISAL 45-49 TULIS [45,49] SESUAI SOAL COK
    
data = [
    [63,67],
    [68,72],
    [73,77],
    [78,82],
    [83,87],
    [88,92]
]
f = [2,3,6,4,4,1]
standarDev(data, f)
```

```python
#STANDAR DEVIASI DATA TUNGGAL

def standarDev(nilai):
    x = sum(nilai) / len(nilai)
    print("rata-rata =", x)
    
    xix = []
    for i in nilai:
        xix.append((i-x)**2)
    print(xix)
    print()
    
    varian = sum(xix) / len(nilai)
    print("Varian (s^2) =",sum(xix),"/",len(nilai),"=", varian)
    print()
    
    sb = math.sqrt(varian)
    print("Standar Deviasi (s) = ", sb)

# KARENA DATA TUNGGAL TIDAK ADA INTERVAL, LANGSUNG TULIS SAJA

data = [
    5,9,7,6,7,8,12,10
]
standarDev(data)
```

```python
# MEDIAN DATA KELOMPOK

def median(data, f):
    cari = []
    n = 1 / 2 * sum(f)
    print("1/2n =", n)
    i = 0
    while sum(cari) < n:
        cari.append(f[i])
        i +=1
    print(cari, i)    
    print("Median ada di interval :", data[i-1])
    bb = data[i-1][0] - 0.5
    print("Batas bawah = ", bb)
    j = 0
    while j < i:
        fmed = sum(f[0:j])
        j+=1
    print("Zf sebelum median =", fmed)
    fmedian = f[i-1]
    print("f median =",fmedian)
    p = data[1][1] - data[0][1]
    print("Panjang kelas = ", p)
    med = bb + ((n - fmed)/fmedian) * p
    print("median =", med)

# PANJANG INTERVAL KELAS MISAL 45-49 TULIS [45,49]

data = [
    [31,40],
    [41,50],
    [51,60],
    [61,70],
    [71,80],
    [81,90],
    [91,100]
]
f = [2,3,5,14,24,20,12]

median(data, f)
```

```python
# MODUS DATA BERKELOMPOK
def modus(data, f):
    print("Letak modus (f terbanyak)", max(f))
    letak = f.index(max(f))
    print("Dengan rentang kelas =", data[letak])
    bb = data[letak][0] - 0.5
    print("batas bawah =", bb)
    d1 = f[letak] - f[letak-1]
    print("d1 =", d1)
    d2 = f[letak] - f[letak+1]
    print("d2 =", d2)
    p = data[1][1] - data[0][1]
    print("Panjang kelas =", p)
    mod = bb + d1/(d1+d2)*p
    print("Modus =", mod)
    
# PANJANG INTERVAL KELAS MISAL 45-49 TULIS [45,49]
    
data = [
    [31,40],
    [41,50],
    [51,60],
    [61,70],
    [71,80],
    [81,90],
    [91,100]
]
f = [2,3,5,14,24,20,12]
modus(data, f)
```

```python
# MEAN DATA BERKELOMPOK
def rata(data, f):
     # menghitung titik tengah
    print("Titik tengah (xi)")
    xi_data = []
    for i in data:
        x = (i[0] + i[1]) / 2
        xi_data.append(x)
    print(xi_data)
    print()

    # menghitung f.xi
    print("f.xi")
    fxi_data = []
    j = 0
    while j < len(data):
        fxi = xi_data[j] * f[j]
        j += 1
        fxi_data.append(fxi)
    print(fxi_data)
    print()

    # menghitung rata2
    print("rata-rata (X)")
    rata = sum(fxi_data) / sum(f)
    print("X =", rata)
    print()
    
# PANJANG INTERVAL KELAS MISAL 45-49 TULIS [45,49] SESUAI SOAL POKONYA
    
data = [
    [31,40],
    [41,50],
    [51,60],
    [61,70],
    [71,80],
    [81,90],
    [91,100]
]
f = [2,3,5,14,24,20,12]
rata(data, f)
```
